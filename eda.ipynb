```python
from datasets import Dataset
from datasets import DatasetDict
from tqdm.auto import tqdm
import os

dataset_dict = {}
for i in tqdm(range(len(data_type))):
    key = data_type[i].split()[-1]
    dataset_dict[key] = Dataset.from_json(data_list[i], num_proc=os.cpu_count())
dataset = DatasetDict(dataset_dict)
dataset
```

# RAG

```python
# Contextual RAG 구현 예시 (간단한 검색 기반 RAG + 임베딩 기반 RAG)
# 모든 데이터 타입(판결문, 법령, 심결례, 유권해석)에서 문장 추출 및 검색

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

import numpy as np

# sentence-transformers 임베딩 모델 사용
from sentence_transformers import SentenceTransformer

# 1. 모든 데이터셋에서 문장만 추출
sentences = []
sentence_sources = []  # 어떤 타입/문서에서 온 문장인지 추적용

for key in dataset.keys():
    for item in dataset[key]:
        # 'sentences' 필드가 리스트라고 가정
        for sent in item['sentences']:
            sentences.append(sent)
            sentence_sources.append(key)  # 타입 정보 저장

# 2. TF-IDF 벡터화
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(sentences)

def retrieve_context_tfidf(query, top_k=3):
    query_vec = vectorizer.transform([query])
    similarities = cosine_similarity(query_vec, tfidf_matrix).flatten()
    top_indices = similarities.argsort()[-top_k:][::-1]
    return [(sentences[i], sentence_sources[i]) for i in top_indices]

# 3. 임베딩 벡터화 (sentence-transformers)
embedding_model = SentenceTransformer("jhgan/ko-sroberta-multitask")  # 한국어 지원 모델
sentence_embeddings = embedding_model.encode(sentences, convert_to_numpy=True)

def retrieve_context_embedding(query, top_k=3):
    query_emb = embedding_model.encode([query], convert_to_numpy=True)
    similarities = cosine_similarity(query_emb, sentence_embeddings).flatten()
    top_indices = similarities.argsort()[-top_k:][::-1]
    return [(sentences[i], sentence_sources[i]) for i in top_indices]

# 4. 예시 질의
query = "계약 해지에 관한 판례를 알려줘"

print("=== Query ===")
print(query)

print("\n=== [TF-IDF 기반] Retrieved Contexts ===")
contexts_tfidf = retrieve_context_tfidf(query, top_k=5)
for i, (ctx, src) in enumerate(contexts_tfidf, 1):
    print(f"{i}. ({src}) {ctx}")

print("\n=== [임베딩 기반] Retrieved Contexts ===")
contexts_emb = retrieve_context_embedding(query, top_k=5)
for i, (ctx, src) in enumerate(contexts_emb, 1):
    print(f"{i}. ({src}) {ctx}")
```
